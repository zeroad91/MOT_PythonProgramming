```python
# Import the library

import matplotlib.pyplot as plt #그림을 그려줘... library.. 
%matplotlib inline  
```

```python
# Create a class Circle

class Circle(object):
    
    # Constructor
    def __init__(self, radius, color='blue'):
        self.radius = radius
        self.color = color 
    
    # Method
    def add_radius(self, r):
        self.radius = self.radius + r
        return(self.radius)
    
    # Method
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.color))
        plt.axis('scaled')
        plt.show()  
```

```python
# Create an object RedCircle

RedCircle = Circle(0, 'red')
print( RedCircle.radius )

RedCircle.add_radius(2)
print(RedCircle.radius)

RedCircle.add_radius(2)
print(RedCircle.radius)

RedCircle.add_radius(2)
print(RedCircle.radius)
```

<h2 id="rect">The Rectangle Class</h2>

```python
# Rectangle Class

class Rectangle(object):
    
    # Constructor #아.. 내가 이제 객체를 하나 초기화 해서 만들어 보겠다. 
    def __init__(self, width = 2, height = 3, color ='r'): #공식처럼 사용하면 됨
        self.height = height 
        self.width = width
        self.color = color
    
    # Method
    def drawRectangle(self):
        plt.gca().add_patch(plt.Rectangle((0, 0), self.width, self.height ,fc=self.color))
        plt.axis('scaled')
        plt.show()
```

```python
SkinnyBlueRectangle = Rectangle(2, 10, 'blue')
SkinnyBlueRectangle.height 
SkinnyBlueRectangle.width
SkinnyBlueRectangle.color
SkinnyBlueRectangle.drawRectangle()
```

```python
# Create a new object rectangle

FatYellowRectangle = Rectangle(20, 5, 'yellow')
```

```python
# Print the object attribute height, width, color

print( FatYellowRectangle.height )
print( FatYellowRectangle.width )
print( FatYellowRectangle.color )
```

```python
FatYellowRectangle.drawRectangle()
```
